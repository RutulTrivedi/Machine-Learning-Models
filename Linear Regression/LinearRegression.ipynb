```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load the dataset
df = pd.read_csv('data.csv')
```

```python
# Basic dataset exploration
print("Dataset Shape:", df.shape)
print("\nDataset Info:")
print(df.info())
print("\nDataset Description:")
print(df.describe())
```

```python
# Extracting features and target variable
x = df['x']
y = df['y']
```

```python
# Reshaping for compatibility with sklearn
x = np.array(x).reshape(-1, 1)
y = np.array(y).reshape(-1, 1)
```

```python
# Scatter plot of the entire dataset
plt.scatter(x, y)
plt.title('Scatter Plot of Whole Data')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Initialize Linear Regression model
model = LinearRegression()
```

```python
# Splitting the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Shape of training and testing data
print('X-train:', X_train.shape)
print('Y-train:', Y_train.shape)
print('X-test:', X_test.shape)
print('Y-test:', Y_test.shape)
```

```python
# Scatter plot of training data
plt.scatter(X_train, Y_train)
plt.title('Scatter Plot of Training Data')
plt.xlabel('X_train')
plt.ylabel('Y_train')
plt.show()
```

```python
# Scatter plot of testing data
plt.scatter(X_test, Y_test)
plt.title('Scatter Plot of Testing Data')
plt.xlabel('X_test')
plt.ylabel('Y_test')
plt.show()
```

```python
# Training the Linear Regression model
model.fit(X_train, Y_train)
```

```python
# Making predictions
Y_predicted = model.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(Y_test, Y_predicted)
r2 = r2_score(Y_test, Y_predicted)

print("\nModel Evaluation:")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R²): {r2}")

# Visualizing predictions vs actual values
plt.scatter(X_test, Y_test, color='blue', label='Actual')
plt.scatter(X_test, Y_predicted, color='red', label='Predicted')
plt.title('Actual vs Predicted Values')
plt.xlabel('X_test')
plt.ylabel('Y_test / Predicted')
plt.legend()
plt.show()
```
