### Importing necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

### Loading the dataset

```python
df = pd.read_csv('data.csv')
```

### Basic dataset exploration

```python
print("Dataset Shape:", df.shape)
print("\nDataset Info:")
print(df.info())
print("\nDataset Description:")
print(df.describe())
```

### Extracting features and target variable

```python
x = df['x']
y = df['y']
```

### Reshaping for compatibility with sklearn

```python
x = np.array(x).reshape(-1, 1)
y = np.array(y).reshape(-1, 1)
```

### Scatter plot of the entire dataset

```python
plt.scatter(x, y)
plt.title('Scatter Plot of Whole Data')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Initialize Linear Regression model

```python
model = LinearRegression()
```

### Splitting the dataset into training and testing sets

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

### Shape of training and testing data

```python
print('X-train:', X_train.shape)
print('Y-train:', Y_train.shape)
print('X-test:', X_test.shape)
print('Y-test:', Y_test.shape)
```

### Scatter plot of training data

```python
plt.scatter(X_train, Y_train)
plt.title('Scatter Plot of Training Data')
plt.xlabel('X_train')
plt.ylabel('Y_train')
plt.show()
```

### Scatter plot of testing data

```python
plt.scatter(X_test, Y_test)
plt.title('Scatter Plot of Testing Data')
plt.xlabel('X_test')
plt.ylabel('Y_test')
plt.show()
```

### Training the Linear Regression model

```python
model.fit(X_train, Y_train)
```

### Making predictions

```python
Y_predicted = model.predict(X_test)
```

### Evaluate the model

```python
mse = mean_squared_error(Y_test, Y_predicted)
r2 = r2_score(Y_test, Y_predicted)

print("\nModel Evaluation:")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R²): {r2}")

# Visualizing predictions vs actual values
plt.scatter(X_test, Y_test, color='blue', label='Actual')
plt.scatter(X_test, Y_predicted, color='red', label='Predicted')
plt.title('Actual vs Predicted Values')
plt.xlabel('X_test')
plt.ylabel('Y_test / Predicted')
plt.legend()
plt.show()
```
